## Testing the model on a webcam

```python
cap = cv2.VideoCapture(0)  ## capture the source 

while cap.isOpened():
    
    _, frame = cap.read()
    
    results = model.predict(source=frame, show=True, conf=0.4)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
